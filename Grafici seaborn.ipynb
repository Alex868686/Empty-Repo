```python

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('https://frenzy86.s3.eu-west-2.amazonaws.com/fav/iris.data', header=None)
df.head()
```

```python
df.columns = ['sepal_len', 'sepal_wid', 'petal_len', 'petal_wid', 'class']
df
```

```python
df.groupby("class").count()
```

```python
import seaborn as sns
sns.pairplot(df)
```

```python
sns.pairplot(df, hue="class",height=2,aspect=1, corner=True)
```

```python
plt.figure(figsize=(10,9))

plt.subplot(2,2,1)
sns.violinplot(df, x='class', y='sepal_len')
plt.subplot(2,2,2)
sns.violinplot(df, x='class', y='sepal_wid')
plt.subplot(2,2,3)
sns.violinplot(df, x='class', y='petal_len')
plt.subplot(2,2,4)
sns.violinplot(df, x='class', y='petal_wid')
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='sepal_wid',data=df,hue='class');
```

```python
sns.jointplot(df, x="petal_len", y="petal_wid", size=6);
```

```python
sns.lmplot(df, x="petal_len", y="sepal_len", hue="class", height=6, aspect=1.2, scatter_kws={"s": 50}, fit_reg=False)
```

```python
plt.figure(figsize=(10,6))
sns.regplot(df, x="sepal_len", y="petal_len", color="green", line_kws={"linewidth": 2,"color":"red","linestyle":"dotted"}, scatter_kws={"s":42})
```

```python
sns.jointplot(df, x='sepal_len',y='petal_len',kind='reg',height=7,ratio=3, color="green", space=0.1)
```

```python
plt.subplots(figsize=(7,5))
sns.heatmap(df.corr(), cmap="coolwarm", linewidths=0.5, xticklabels=False, annot=True) #yticklabels=False, cbar=False
```
